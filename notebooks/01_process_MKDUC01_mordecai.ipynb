```python
from loguru import logger
from os import path
import json
from geo_kpe_multidoc import (
    GEO_KPE_MULTIDOC_DATA_PATH,
    GEO_KPE_MULTIDOC_MORDECAI_ES_URL,
)
from mordecai import Geoparser
```

```python
GEO_KPE_MULTIDOC_MORDECAI_ES_URL = '35.180.192.139'
```

```python
parser = Geoparser(es_hosts=GEO_KPE_MULTIDOC_MORDECAI_ES_URL, progress=False)
```

```python
from elasticsearch_dsl import Search, Q
from elasticsearch import Elasticsearch
```

```python
kwargs = dict(
        hosts=GEO_KPE_MULTIDOC_MORDECAI_ES_URL,
        port=9200,
        use_ssl=False,
        timeout=240,
    )

CLIENT = Elasticsearch(**kwargs)
S = Search(using=CLIENT, index="geonames")
parser.conn = S
```

```python
# parser = Geoparser(es_hosts=GEO_KPE_MULTIDOC_MORDECAI_ES_URL, progress=False, timeout=30)

with open(path.join(GEO_KPE_MULTIDOC_DATA_PATH, "MKDUC01/MKDUC01.json")) as s_json:
    docs = json.load(s_json)

    for doc_group in docs:
        try:
            logger.info(f"Geoparsing group {doc_group}")
            if doc_group not in [
                # # "d04",
                # # "d05",
                # # "d06",
                # "d08", # error
                # # "d11",
                # # "d12",
                # # "d13",
                # "d14", # error 
                # # "d15",
                # # "d19",
                # "d22", # error
                # # "d24",
                # # "d27",
                # "d28", # error
                # # "d30",
                # # "d31",
                # # "d32",
                # "d34", # error
                # # "d37",
                # # "d39",
                # # "d41",
                # "d50", # error
                # "d53", # error
                "d59", # error
            ]:
                continue

            res = {
                doc_group: {
                    doc_id: str(parser.geoparse(text))
                    for doc_id, text in docs[doc_group]["documents"].items()
                }
            }

            with open(
                path.join(
                    GEO_KPE_MULTIDOC_DATA_PATH,
                    f"MKDUC01/MKDUC01-mordecai-{doc_group}.json",
                ),
                mode="w",
            ) as f:
                logger.info(f"Geoparsing save {doc_group}")
                json.dump(res, f, indent=4, separators=(",", ": "))
            
            del res
            parser.query_geonames.cache_clear()
            parser.query_geonames_country.cache_clear()

        except Exception as e:
            logger.error(e)
            with open(
                path.join(
                    GEO_KPE_MULTIDOC_DATA_PATH,
                    "MKDUC01/MKDUC01-mordecai-errors.txt",
                ),
                mode="a",
            ) as f:
                f.write("\n")
                f.write(doc_group)
```
